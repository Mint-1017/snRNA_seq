```python
#PTS1 file
PTS1 = []

meta = []
Ext_meta = "/path/to/Extract_control_GSE139107_MouseIRI.metadata.txt"

Ext_meta_f = open(Ext_meta,'r',encoding='utf-8')
for line in Ext_meta_f:
    meta.append(line.split())

for line in meta:
    if line[3] == 'PTS1':
        PTS1.append(line[0]) 
        
f = open('SpeedTest_ctrl_PTS1.txt','w')

for line in PTS1:
    f.write(line)
    f.write('\t')
f.write('\n')

ctrl = "/path/to/GSE139107_MouseIRI_control.dge.txt"
ctrl_f = open(ctrl,'r',encoding='utf-8')

l = []
for line in ctrl_f:
    l.append(line.split())


for j in range(1,27134):
    f.write(l[j][0])
    f.write('\t')
    
    k = 0
    for line in PTS1:
        for i in range(k,26654): 
            if line == l[0][i]:
                f.write(l[j][i+1])
                f.write('\t')
                k = i
                break
    f.write('\n')
    
f.close()
```
