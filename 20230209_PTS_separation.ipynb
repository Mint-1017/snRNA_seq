```python
def dofunc (metadata,inputpath,outputpath,PTS_word):
    meta = []
    P = []
    Ext_meta = metadata
    Ext_meta_f = open(Ext_meta,'r',encoding='utf-8')
    
    for line in Ext_meta_f:
        meta.append(line.split())
    for line in meta:
        if line[3] == PTS_word:
            P.append(line[0]) 
        
    f = open(outputpath,'w')
    for line in P:
        f.write(line)
        f.write('\t')
    f.write('\n')
    
    data = inputpath
    data_f = open(data,'r',encoding='utf-8')
    
    l = []
    for line in data_f:
        l.append(line.split())
        
    for j in range(1,len(l)):
        f.write(l[j][0])
        f.write('\t')
        k = 0
        for line in P:
            for i in range(k,len(meta)): 
                if line == l[0][i]:
                    f.write(l[j][i+1])
                    f.write('\t')
                    k = i
                    break
        f.write('\n')
    
    f.close()
```
